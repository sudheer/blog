## Spark Query Plan



Spark Query Plan Optimization

- **Analysis** : analyses where datasets are coming from, what are columns and datatypes.
- **Logical Optimization**: optimization of logical plan.
  1. filter predicate pushdown
  2. Column Pruning
- **Physical Planning**: physical planning
  1. choose between broadcast jon and shuffle joins to reduce network traffic.
  2. Determine how to shuffle data, how to partition and ordering data etc.
- **Code Generation**: Compile parts of query to java bytecode

## Dataframe vs Datasets

| | DataFrames | Datasets |
|--|--|--|
| type | Untyped API | Typed API |
| Syntax Errors | Compile Time | Compile Time |
| Analysis Errors | Runtime | Compile Time |

In Short

**DataFrame** : Abstration on top of data which gives columns and datatypes. Downside is type safety as we need to refer attributes by String names and doesn't have compile time safetly. 

**DataSet** : Extension of dataframe API, has compile time safety. case class is used to define the structure of data schema in Dataset. Encoders are used to handle conversion between JVM objects to binary representation.
